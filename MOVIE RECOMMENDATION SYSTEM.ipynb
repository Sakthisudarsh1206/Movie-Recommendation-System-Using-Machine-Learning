```python
movies=movies.merge(credits,on="title")
movies.shape
#we merge the two datasets with the common column-title
```

```python
movies=movies[['movie_id','genres','keywords','title','cast','crew','overview']]
```

```python
movies.head()
#updated movies table with only the needed columns
```

```python
movies.isnull().sum()
#check for null values
```

```python
movies.dropna(inplace=True)
#remove empty rows
```

```python
movies.duplicated().sum()
#check for duplicate values
```

```python
movies.iloc[0].genres
#we see the data is in the form of list of dictionaries
```

```python
#for genres column
def conv(obj):
    list=[]
    for i in ast.literal_eval(obj):
        #we convert string of lists to list using ast
        list.append(i['name'])
    return list

#for cast column
def conv3(obj):
    list=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=4:
            list.append(i['name'])
            count+=1
    return list

#for crew column
def fetch_d(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if (i['job']== 'Director'):
            l.append(i['name'])
    return l

```

```python
#movies.iloc[0].cast
```

```python
#movies.iloc[0].crew
```

```python
movies['genres']=movies['genres'].apply(conv)
#convert crew columns
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(conv)
movies.head()
```

```python
movies['cast']=movies['cast'].apply(conv3)
movies.head()
```

```python
movies['crew']=movies['crew'].apply(fetch_d)
#convert genres columns
movies.head()
```

```python
#converting paragraph words to single words in list
movies['overview']=movies['overview'].apply(lambda x: x.split(" "))
movies.head()
```

```python
#remove space in each column values
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
movies['overview']=movies['overview'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
#create new column label with all attributes/tags
movies['label']=movies['genres']+movies['keywords']+movies['cast']+movies['crew']+movies['overview']
movies['label']
```

```python
df=movies[['movie_id','title','label']]
print(df)
```

```python
#converting list to string
df['label']=df['label'].apply(lambda x:" ".join(x))
df.head()
```

```python
df['label']=df['label'].apply(lambda x:x.lower())
df.head()
```

```python
#convert text to vectors with 5000 vectors and exclude stop words like a,an,the.. etc
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors= cv.fit_transform(df['label']).toarray()
```

```python
#get the features or words 
cv.get_feature_names_out()
```

```python
import nltk
```

```python
#we use porter stemmer algorithm here to remove the suffixes from words so that same words with diff tenses can be combined
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#helper function

def st(obj):
    l=[]
    for i in obj.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
df['label']=df['label'].apply(st)
```

```python
df['label'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi= cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index= df[df['title']== movie].index[0]
    dist=simi[movie_index]
    movies_list=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(df.iloc[i[0]].title)
        
```

```python
recommend('Mission: Impossible')
```
